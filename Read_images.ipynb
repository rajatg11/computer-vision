```python
#import/read the image
img = cv2.imread("opencv_image.png")
```

We can use **matplotlib** library to display the image in the Jupyter Notebook iteself

```python
import matplotlib.pyplot as plt
```

```python
plt.imshow(img)
```

The above image is different from the Original Image. Blue color and Red Color position is interchanged because OpenCV reads image in BGR format but Matplotlib expect image to be in RGB format.

So, we have to convert the color of the image and display it again.

```python
imgRGB = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(imgRGB)
```

The Above image looks similar to our Original Image
